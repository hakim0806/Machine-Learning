## Section 8 — Categorical Feature Exploration

Look at distributions for each categorical feature.


```python
# TODO:
# For each categorical column:
# 1. Plot a bar chart of value_counts()
# 2. Observe category balance and potential issues
for c in categorical_cols:
  plt.figure(figsize=(5,4))
  df[c].value_counts().plot(kind="bar")
  plt.title(f"Value counts for {c}")
  plt.ylabel("Count")
  plt.tight_layout()
  plt.show()
```

```python
cols = categorical_cols
n = len(cols)

for i in range(0, n, 3):               # প্রতি 3 করে এগোবে
    row_cols = cols[i:i+3]             # এক row এর 3টি কলাম নেবে

    plt.figure(figsize=(15, 4))        # এক row এর figure

    for j, c in enumerate(row_cols):
        plt.subplot(1, 3, j+1)         # 1 row, 3 column layout
        df[c].value_counts().plot(kind="bar")
        plt.title(f"Value counts for {c}")
        plt.ylabel("Count")

    plt.tight_layout()
    plt.show()
```

## Section 11 — Categorical Features versus Target (Crosstab and Stacked Bar Plot)

Examine how target proportions change across categories.


```python
# TODO:
# For each categorical feature:
# 1. Create a crosstab between the feature and target with normalize="index"
# 2. Plot a stacked bar chart of the proportions
# 3. Comment on how the target distribution changes across categories

#Categorical Features vs Target
for c in categorical_cols:
  ct = pd.crosstab(df[c], df[target_col], normalize="index")
  print(f"\nProportion of Is froud ditection within {c}")
  display(ct)

  ct.plot(kind="bar", stacked=True, figsize=(6,4))
  plt.title(f"Froud ditection proportion by {c}")
  plt.ylabel("proportion")
  plt.xticks(rotation=0)
  plt.tight_layout()
  plt.show()
```
